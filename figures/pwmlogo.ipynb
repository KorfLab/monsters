```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
plt.ion()

import logomaker as lm
```

```python
donor = pd.read_csv('../data/donor.pwm', sep=' ', skiprows=1, header=None)
donor.drop(donor.columns[-1], axis=1, inplace=True)
donor.columns = ['A', 'C', 'G', 'T']
acceptor = pd.read_csv('../data/acceptor.pwm', sep=' ', skiprows=1, header=None)
acceptor.drop(acceptor.columns[-1], axis=1, inplace=True)
acceptor.columns = ['A', 'C', 'G', 'T']
```

```python
donor = lm.validate_matrix(donor)
acceptor = lm.validate_matrix(acceptor)
```

```python
logo = lm.Logo(donor, font_name='Arial Rounded MT Bold')
logo.ax.set_xlabel('Position',fontsize=14)
plt.savefig('donor.png')
```

```python
logo = lm.Logo(acceptor, font_name='Arial Rounded MT Bold')
logo.ax.set_xlabel('Position',fontsize=14)
plt.savefig('acceptor.png')
```
